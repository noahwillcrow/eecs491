## The Problem

As in exercise one, I want to explore the likely cause of a fire alarm. However, now, I want to turn the time that it takes for us to be allowed back in to be represented in finer detail than "quick" or not. I'll ignore the other two variables for now. I'll now change "quick" to be "time", and it will be represented as a continuous value in seconds. I will say that time = 0 is the moment that everyone has vacated my building.

## Defining Probabilities

If the cause is a drill, then the resulting value for time can be represented as a half normal with mean 0s and standard deviation 30s. However if the cause is a small mistake, such as a microwave mishap, then the value of time can be represented as a normal distribution with mean 50s and standard deviation 30s. Finally, if the cause is an actual fire, then the time can be represented as a normal distribution with mean 260s and standard deviation 70s.

And, as before, these are my probabilities:
- $P(Mistake) = 0.75$
- $P(Drill) = 0.35$
- $P(Fire) = 0.05$

## Deriving Probabilities from Theory

In order to actually derive the probabilities of each cause, I will need to allow for an epsilon value around the given time t. I will set this epsilon to 2 seconds, giving a four second range for every given value of t.

Given a time t:
- $P(Drill \mid t - \epsilon \leq T \leq t + \epsilon) = P(D \mid T \leq t + \epsilon) = \alpha \dfrac{\sqrt{2}}{\sigma \sqrt{\pi}}exp(- \dfrac{(t + \epsilon)^2}{2\sigma^2})$
  - $P(Drill \mid t - \epsilon \leq T \leq t + \epsilon) = P(D \mid T \leq t + \epsilon)$ because $P(Drill \mid T < 0) = 0$
- $P(Mistake \mid t - \epsilon \leq T \leq t + \epsilon) = \alpha (\Phi(\dfrac{t + \epsilon - \mu}{\sigma}) - \Phi(\dfrac{t - \epsilon - \mu}{\sigma}))$
- $P(Fire \mid t - \epsilon \leq T \leq t + \epsilon) = \alpha \dfrac{\sqrt{2}}{\sigma \sqrt{\pi}}(exp(- \dfrac{(t + \epsilon)^2}{2\sigma^2}) - exp(- \dfrac{(t - \epsilon)^2}{2\sigma^2}))$

## Creating the Model in PyMC3

In order to put this into code, I will use PyMC3, which thankfully comes pre-packaged with all three of the above distributions.

```python
import numpy
import pymc3
import theano

p_mistake = 0.75
p_drill = 0.35
p_fire = 0.05

mistake_mean = 50
mistake_sd = 30
drill_sd = 30
fire_mean = 260
fire_sd = 70

time_epsilon = 2

num_observations = 2500

observed_data = theano.shared(numpy.zeros(num_observations))

with pymc3.Model() as model:
    mistake_happened = pymc3.Bernoulli('Mistake', p = p_mistake)
    drill_happened = pymc3.Bernoulli('Drill', p = p_drill)
    fire_happened = pymc3.Bernoulli('Fire', p = p_fire)
    
    time_from_mistake = pymc3.Normal('Mistake_Time', mu = mistake_mean, sd = mistake_sd)
    time_from_drill = pymc3.HalfNormal('Drill_Time', sd = drill_sd)
    time_from_fire = pymc3.Normal('Fire_Time', mu = fire_mean, sd = fire_sd)
    
    time = pymc3.Normal('Time', mu = mistake_happened*time_from_mistake + drill_happened*time_from_drill + fire_happened*time_from_fire, sd = time_epsilon)
```

## Generating Data

In order for this to be useful, I need to generate data on the model and show the results. This can be done using PyMC3's tracing capabilities.

```python
%matplotlib inline

with model:
    trace = pymc3.sample(num_observations, chains = 2)
    pymc3.traceplot(trace)
```
