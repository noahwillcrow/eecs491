### Utility functions

Now we need a function to move this data into the model we have, which needs all data points to be in the range [-0.5, 0.5]

```python
import numpy as np

def audionorm(data):
    # ensure data is ndarray with float numbers
    data = np.asarray(data).astype('float')
    # calculate lower and upper bound
    lbound, ubound = np.min(data), np.max(data)
    if lbound == ubound:
        offset = lbound
        scalar = 1
        data = np.zeros(size=data.shape)
    else:
        offset = (lbound + ubound) / 2
        scalar = 1 / (ubound - lbound)
        data = (data - offset) * scalar
    # return normalized data
    return data
```

And then we need a function to mix our two audio files together

```python
def simpleMixer(S):
    nchannel = S.shape[0]
    # generate a random matrix
    A = np.random.uniform(size = (nchannel,nchannel))
    # generate mixed audio data
    X = A.dot(S)
    
    return X, A
```

And also a utility function for drawing some data and a mixing matrix.

```python
%matplotlib inline

import matplotlib.pyplot as plt

def drawDataWithMixingMatrix(data, mat):
    # plot data points
    plt.scatter(data[0], data[1], s=1)
    # calculate axis length
    lenAxis = np.sqrt(np.sum(np.square(mat), axis=0))
    # calculate scale for illustration
    scale = np.min(np.max(np.abs(data), axis=1) / lenAxis.T)
    # draw axis as arrow
    plt.arrow(0, 0, scale * mat[0,0], scale * mat[1,0], shape='full', color='r')
    plt.arrow(0, 0, scale * mat[0,1], scale * mat[1,1], shape='full', color='r')
```

And last but not least for the utility functions, a function for comparing two mixing matrices on the same data.

```python
def compareMixingMatrix(data, matA, matB):
    plt.figure(figsize=(16, 8))
    # plot first mixing matrix
    plt.subplot(1,2,1)
    drawDataWithMixingMatrix(data, matA)
    # plot first mixing matrix
    plt.subplot(1,2,2)
    drawDataWithMixingMatrix(data, matB)
```

### Independent Component Analysis and Blind Source Separation

Now begins the key to this project&mdash;ICA and BSS. The point here is to take raw audio data, $X$, and blindly find the independent sources of the audio using probability theory. The idea is to iteratively find the gradient of the log probability of the current estimated sources matrix, $S$, and use that to find the gradient of the mixing matrix, $A$. We can find $S$ with the following formula:

$$
S = A^{-1}X
$$

Initially, we will start with $A = I_n$ so that we just start with the log likelihood of the raw data itself.

For this log likelihood, we will assume that each data point is distributed according to the Laplacian distribution. We can then derive the log probability of each point like so:

$$
\begin{align}
log(P(S_i))^\prime & = log(\dfrac{1}{2b} \exp{\dfrac{-\left|S_i - \mu\right|}{b}})^\prime \\
& = \dfrac{d}{dS_i} (log(\dfrac{1}{2b}) + \dfrac{-\left|S_i - \mu\right|}{b}) \\
& = 0 + \dfrac{\mu - S_i}{b\left|\mu - S_i\right|} \\
& = \dfrac{1}{b}sign(\mu - S_i)
\end{align}
$$

$\dfrac{1}{b}$ is just some constant scalar value, so we can ignore that in our computations. And, also, since all of our data is centered, we get that $\mu = 0$. So we ultimately get:

$$
\begin{align}
log(P(S_i))^\prime & \propto sign(\mu - S_i) \\
& \propto sign(0 - S_i) \\
& \propto -sign(S_i)
\end{align}
$$

So, each iteration, to compute the gradient matrix, we will use the following formula:

$$
\begin{align}
Z & = log(P(S))^\prime \\
& = -sign(S) \\
\Delta A & = -A(ZS^T + I_n)
\end{align}
$$

And we can then compute our new mixing matrix, $A^\prime$:

$$
A^\prime = A + \alpha \Delta A
$$

where $\alpha \in \mathbb{R}$. $\alpha$ acts as the step size for our iterative algorithm.

And now let's put it into code!

The first step is to import two key libraries&mdash;numpy's linear algebra library and matplotlib's pyplot.

```python
from numpy import linalg as la
from matplotlib import pyplot as plt
```

Now I want to create a function that "unmixes" my data according to a mixing matrix. That is, I want a function to calculate S according to this function:

$$
S = A^{-1}X
$$

```python
def unmix_data(mixing_matrix, data):
    # S = A^(-1)X
    s = np.matmul(la.inv(mixing_matrix), data)
    return s
```

And now I want another quick function. This one will calculate Z as found above. That is, it will calculate the result of this:

$$
Z = -sign(S)
$$

```python
def calc_z(s):
    # Z = d(log(P(S)))/dx = -sign(S) [see above]
    z = -np.sign(s)
    return z
```

With those two quick calculation methods, I can create a function to calculate my gradient for the mixing matrix. This is the function that actually implements the key function of this entire project:

$$
\Delta A = -A(ZS^T + I_n)
$$

```python
def calc_gradient_matrix(mixing_matrix, s):
    z = calc_z(s)
    zs_t = np.matmul(z, s.T) # ZS^T
    identity_matrix = np.identity(len(zs_t)) # I_n
    gradient_matrix = (-1) * np.matmul(mixing_matrix, zs_t + identity_matrix) # A(ZS^T + I)
        
    return gradient_matrix
```

Finally, I can create my ICA function. It will iteratively calculate a gradient and add a scalar multiple of that gradient to the mixing matrix until it reaches its end condition&mdash;either the gradient's Frobenius norm drops below a threshold or the loop hits too many iterations.

```python
def ica(data, step_size, convergence_min, max_iterations):
    mixing_matrix = np.identity(len(data)) # Initial value of A = I_n
    cur_data = unmix_data(mixing_matrix, data) # Initial value of S
    
    grad_norms = [] # For graphing
    
    cur_grad_norm = convergence_min + 1 # Set this to the min + 1 to essentially create a do-while loop
    
    n_iter = 0 # Just to keep from running forever
    
    # Consider convergence to be when the Frobenius norm of the gradient drops below a threshold
    while (cur_grad_norm > convergence_min and n_iter < max_iterations):
        gradient_matrix = calc_gradient_matrix(mixing_matrix, cur_data)
        mixing_matrix += step_size * gradient_matrix # a * dA
        
        cur_data = unmix_data(mixing_matrix, data) # Update current S value
        cur_grad_norm = la.norm(gradient_matrix) # Update gradient norm to check for convergence
        grad_norms.append(cur_grad_norm)
        
        n_iter += 1

    return mixing_matrix, cur_data, grad_norms
```

For analysis purposes, I want to be able to call the ICA and plot its gradient progress in one line. I'll call this function the BSS function, as the rest of my code can treat the BSS as a black box that just separates sources. The fact that BSS uses ICA should not matter to the rest of my code. I will also have my BSS function take an argument for whether it should plot the progress the gradient makes as the ICA takes place.

```python
def plot_gradient_progress(gradient_progress, goal_value):
    plt.plot(gradient_progress, label = "Gradient magnitude")
    plt.plot([0, len(gradient_progress)-1], [goal_value, goal_value], label = "Goal magnitude")
    plt.xlabel("Iterations")
    plt.ylabel("Gradient magnitude")
    plt.title("Gradient progress")
    plt.legend()
    plt.show()
    
def bss(data,step_size = 0.1, convergence_min = 50, max_iterations = 1000,
        should_plot_gradient_progress = False):
    data = np.asarray(data).astype("float")

    mixing_matrix, recovered_data, gradient_progress = ica(data, step_size, convergence_min, max_iterations)
    
    if (should_plot_gradient_progress):
        plot_gradient_progress(gradient_progress, convergence_min)
    
    return mixing_matrix, recovered_data
```

That looks good! Now we can analyze each component of the estimated mixing matrix and compare it with the basis vector for our music "subspace" and see if it falls close enough.

To do this, I'll want a function.

```python
# onb stands for orthonormal basis
def does_any_component_project_well_into_vector_space(component_matrix, vector_space_onb, min_norm_ratio = 0.95):
    for component in component_matrix.T:
        projected_vector = np.zeros_like(component)
        for onb_vector in vector_space_onb.T:
            projected_vector += np.dot(component, onb_vector) * onb_vector
            
        projection_norm_ratio = la.norm(projected_vector) / la.norm(component)
        if projection_norm_ratio >= min_norm_ratio:
            return True
    return False
```

And now I can use that!

```python
drum_onb = np.asarray([[1.], [1.]])
drum_onb /= la.norm(drum_onb)

is_music = does_any_component_project_well_into_vector_space(music_A, drum_onb)
print("And the verdict is:")
if is_music:
    print("The sound track contained music (correct!)")
else:
    print("The sound track did not contain music (incorrect)")
```
