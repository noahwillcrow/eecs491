## Derivation of EM

### Expectation Step
This is the step in which we compute the expectation for a given set of parameters.

The probability that the n<sup>th</sup> element of data is in class $k$ is given below:

$$
p_{n,k} = p(c_k \mid  x^{(n)}, \theta_{1:K})
$$

Then the probability of a given class, $p_k$ is just the sum over n of $p_{n,k}$.

So what exactly is $p(c_k \mid  x^{(n)}, \theta_{1:K})$?

$$
\begin{align}
p(c_k \mid  x^{(n)}, \theta_{1:K}) & = \dfrac{p(x^{(n)} \mid c_k, \theta_{1:K})p(c_k)}{p(x^{(n)})} \\
& = \dfrac{p(x^{(n)} \mid c_k, \theta_{1:K})p(c_k)}{\sum_k{p(x^{(n)} \mid c_k, \theta_{1:K})p(c_k)}}
\end{align}
$$

We can then use the normal distribution of class k and use that function to determine the probability that x<sup>(n)</sup> is in that class:

$$
\begin{align}
p(c_k \mid  x^{(n)}, \theta_{1:K})
& = \dfrac{\mathcal{N}(x^{(n)} \mid \theta_{1:K})p(c_k)}{\sum_k{\mathcal{N}(x^{(n)} \mid \theta_{1:K})p(c_k)}}
\end{align}
$$

### Maximization Step

For this, we will want to use log probabilities to avoid division by zero.

$$
\begin{align}
log(p(c_k \mid x^{(n)}, \theta_{1:K}))
& = log(\mathcal{N}(x^{(n)} \mid c_k, \theta_{1:K})p(c_k)) - log(\sum_k{\mathcal{N}(x^{(n)} \mid c_k, \theta_{1:K})p(c_k)})
\end{align}
$$

So now we want to maximize the mean, covariance, and class priors by iteration. These new values will be denoted with a $\prime$ symbol.

$$
\begin{align}
log(\mu_k^\prime) & = \sum_n{log((p_{n,k})x^{(n)}) - log(p_k)} \\
log(\Sigma_k^\prime) & = \sum_n{log((p_{n,k})(x^{(n)} - \mu_k)(x^{(n)} - \mu_k)^T) - log(p_k)} \\
log(p_k^\prime) & = log(p_k)
\end{align}
$$

```python
def plotGaussianModel2D(mu, sigma, pltopt='k'):
    if sigma.any():
        # calculate ellipse constants
        c = chi2.ppf(0.9, 2) # use confidence interval 0.9
        # get eigen vector and eigen values
        eigenValue, eigenVector = np.linalg.eig(sigma)
        # calculate points on ellipse
        t = np.linspace(0, 2*np.pi, 100) # draw 100 points
        u = [np.cos(t), np.sin(t)]
        w = c * eigenVector.dot(np.diag(np.sqrt(eigenValue)).dot(u))
        z = w.T + mu
    else:
        z = mu
    # plot ellipse by connecting sample points on curve
    plt.plot(z[:,0], z[:,1], pltopt)
    
def colorPicker(index):
    colors = 'rgbcmyk'
    return colors[np.remainder(index, len(colors))]

def gmmplot(data, gmm):
    # plot data points
    plt.scatter(data[:, 0], data[:, 1], s=4)
    # plot Gaussian model
    color = 'rgb'
    for index, model in enumerate(gmm):
        plotGaussianModel2D(model['mean'], model['covariance'], colorPicker(index))
```

### Implementing the EM Algorithm

In order to actually see the results of the EM derivations above, they need to be put into code. The first step of doing that is to create the expectation step function. To do this, I created another function, `gather_responsibilities`, which finds the responsibilities each class has for each data point. It seemed cleaner to do this in a separate function from the `expectation` function. The `expectation` function then calls `gather_responsibilities` for each data point and adds those responsibilities to a larger array which is actually used to generate the expectation results.

```python
from scipy.stats import multivariate_normal as mvg

def get_responsibility(x, prior, mean, covar):
    mvg_pdf = mvg.pdf(x, mean = mean, cov = covar)
    return mvg_pdf * prior

def gather_responsibilities(x, m):
    responsibilities = []
    
    # Calculate the likelihood that class k is responsible for generating data point x
    for k in range(len(m)):
        prior = m[k]['prior']
        mean = m[k]['mean']
        covar = m[k]['covariance']
        responsibilities.append(get_responsibility(x, prior, mean, covar))
        
    # Normalize!
    divisor = np.sum(responsibilities)
    responsibilities /= divisor
    
    return responsibilities

def expectation(d, m):
    responsibilities = []
    for n in range(len(d)):
        # First, iterate through all values and compute the likelihoods / responsibilities
        resps = gather_responsibilities(d[n], m)
        responsibilities.append(resps)
        
    # And now normalize the results.
    divisors = np.array(responsibilities).sum(axis=0)
    for resp in responsibilities:
        for k in range(len(divisors)):
            resp[k] /= divisors[k]
        
    return responsibilities
```

With expectation taken care of, I now need to use those responsibilities to maximize my model's parameters. To keep things simple at first, I will simply maximize the means for each class in a `maximization_mean` method.

```python
def maximization_mean(p, d, m):
    new_gmm = []
    
    for k in range(len(m)):
        new_mean = np.zeros_like(m[k]['mean'])
        
        for n in range(len(d)):
            x = d[n]
            new_mean += p[n][k] * x
            
        new_gmm.append({'mean': new_mean, 'covariance': m[k]['covariance'], 'prior': m[k]['prior']})
        
    return new_gmm
```

And now I can actually graph the results of this!

```python
# make a true copy of our model
gmmcp = copy.deepcopy(gmm)

# create figure
plt.figure(figsize=(16, 8))
# improve model with EM-Algorithm
for i in range(5):
    # plot current status
    plt.subplot(231 + i)
    gmmplot(data, gmmcp)
    # excute EM-Algorithm
    for j in range(5):
        posterior = expectation(data, gmmcp)
        gmmcp = maximization_mean(posterior, data, gmmcp)
# plot final status
plt.subplot(236)
gmmplot(data, gmmcp)
```

# Exercise 2

So maximizing the mean alone gave good results. But the covariance matrix also needs to get some work done to really zero in on the clusters. So now I'll create a more `maximization` function that maximizes the mean, covariance, and prior values for each class.

```python
def maximization(p, d, m):
    new_gmm = []
    
    for k in range(len(m)):
        cur_mean = m[k]['mean']
        new_mean = np.zeros_like(m[k]['mean'])
        new_covar = np.zeros_like(m[k]['covariance'])
        new_prior = 0
        
        for n in range(len(d)):
            new_mean += p[n][k] * d[n]
            new_prior += p[n][k] / np.sum(p[n])
            
        #Have to do covariance after computing the new mean
        for n in range(len(d)):
            x_minus_mean = np.reshape(d[n] - new_mean, (len(d[n]), 1))
            x_minus_mean_T = x_minus_mean.transpose()
            mat_prod = p[n][k] * (x_minus_mean * x_minus_mean_T)
            new_covar += mat_prod
            
        new_prior /= len(d)
            
        new_gmm.append({'mean': new_mean, 'covariance': new_covar, 'prior': new_prior})
        
    return new_gmm
```

And now to see how much better this method works than the `maximization_mean` function did.

```python
# make a true copy of our model
gmmcp = copy.deepcopy(gmm)

# create figure
plt.figure(figsize=(16, 8))
# improve model with EM-Algorithm
for i in range(5):
    # plot current status
    plt.subplot(231 + i)
    gmmplot(data, gmmcp)
    # excute EM-Algorithm
    for j in range(1):
        posterior = expectation(data, gmmcp)
        gmmcp = maximization(posterior, data, gmmcp)
# plot final status
plt.subplot(236)
gmmplot(data,gmmcp)
```

### Using Principal Component Analysis

Principal-component analysis (PCA) is technique used to identify the most important dimensions to distinguish a data set. This methodology is based upon eigenspectrums and matrix algebra. To put it simply, the analysis finds N dimensions in the data that can best be used to distinguish classes in the data, allowing us to dramatically reduce the space of the problem.

Implementing this in code can be pretty simple:

```python
from sklearn.decomposition import PCA

def dimReducePCA(X, ndim):
    pca = PCA(n_components = ndim)
    pca.fit(X)
    X -= np.mean(X, axis=0)
    covar = np.matmul(X.T, X) / X.shape[0]
    
    eigenvalues, eigenvectors = np.linalg.eig(covar)
    
    return pca.transform(X), eigenvectors, eigenvalues
```

And now, let's use PCA to identify the top-two components for the MNIST dataset.

```python
# reduce the dimension of data to 2 using PCA
mnistData2D, eigenVectors, eigenValues = dimReducePCA(mnistData, 2)
```

```python
# randomly pick coordinates as center of Gaussian distribution
def randinitMean(n, drange):
    mu = []
    for m in range(n):
        mu.append([np.random.uniform(drange[i]['MIN'], drange[i]['MAX']) for i in range(len(drange))])
    # return inf form of numpy.ndarray
    return np.asarray(mu)
```

```python
def randinitCovariance(n):
    sigma = []
    for m in range(n):
        # randomly pick a direction
        angle = np.random.uniform(0, 2 * np.pi)
        # create eigen vector with the direction and direction orthogonal to it
        eigenVector = np.asarray([[np.cos(angle), np.sin(angle)],
                                  [np.cos(angle + np.pi/2), np.sin(angle + np.pi/2)]])
        # randomly generate eigen values
        eigenValue = abs(np.random.normal(size=2))
        # compose covariance matrix
        sigma.append(eigenVector.T.dot(np.diag(eigenValue).dot(eigenVector)))
    # return in form of numpy.ndarray
    return np.asarray(sigma)
```

# Exercise 5 (Bonus)

So all of this has been great. But two pixels is hardly going to tell us anything about the number they represent. So why not amp it up? Why not go with 28? Or more?

My plan here is to run the EM algorithm on a larger number of principal components, then simply grab the most likely element for each Gaussian mixture and render it's underlying image. If all goes well, we should see better results as we go up in dimensions.

The first step to this is to create a function to grab the most likely element of a class.

```python
def get_most_likely_element_index(d, prior, mean, covar):
    index = -1
    resp = 0
    
    for n in range(len(d)):
        r = get_responsibility(d[n], prior, mean, covar)
        if r > resp:
            index = n
            resp = r
            
    if index == -1:
        print("No elements found for class")
        return
    
    return index

def get_most_likely_element_indices_per_class(d, m):
    indices = []
    for k in range(len(m)):
        indices.append(get_most_likely_element_index(
            d, m[k]['prior'], m[k]['mean'], m[k]['covariance']
        ))
    return indices
```

And now I can simply run my PCA/EM combo over a multitude of dimension sizes and show the results! Since I'll be doing this repetitively, I'll make another function so I can call this with one line. One issue, though, is that I need a new method to create my initial covariance, as the one above is only capable of producing 2x2 matrices. Let's do that first.

```python
def create_random_covariance_matrices(n, dim, data_range):
    matrices = []
    for i in range(n):
        diag_entries = [np.random.uniform(max(data_range[j]['MIN'], 1), max(data_range[j]['MAX'], 5)) for j in range(dim)]
        covar_matrix = np.diag(diag_entries)
        matrices.append(np.matmul(covar_matrix, covar_matrix))
    return matrices
```

And now I can create the main method of the section!

```python
def show_most_likely_elements_for_pca_size(d, n_pcs, n_classes = 10, n_em_iters = 25, n_data = 1000):    
    # run the PCA
    reduced_data, eigenvectors, eigenvalues = dimReducePCA(d, n_pcs)
    
    # randomly pick a subset of data
    data_set = reduced_data[np.random.randint(len(reduced_data), size=n_data)]
    
    # get the gmm
    data_range = []
    for i in range(data_set.shape[1]):
        data_range.append({'MIN': min(data_set[:,i]), 'MAX': max(data_set[:,i])})
    means = randinitMean(n_classes, data_range)
    covariances = create_random_covariance_matrices(n_classes, n_pcs, data_range)
    m = [{'mean': means[k], 'covariance': covariances[k], 'prior': 1.0/n_classes} for k in range(n_classes)]
    
    # run the EM
    for i in range(n_em_iters):
        posterior = expectation(data_set, m)
        m = maximization(posterior, data_set, m)
        
    # get most likely elements
    most_likely_indices = get_most_likely_element_indices_per_class(data_set, m)
    
    # plot it
    imageset = [d[n] for n in most_likely_indices]
    showImageSamples(imageset, n_classes, random=False)
```
